```
import numpy as np
from RL_for_reco.finiteMDP_for_Reco import Multiple_FMDP_Reco
```

```
p = np.random.random_sample([3, 5, 3])
r = np.random.choice([-1, 0, 1], (3, 5, 3))
for i in range(p.shape[0]):
  for j in range(p.shape[1]):
    p[i, j] = p[i, j] / sum(p[i, j])
```

```
n_step = 1000
n_episode = 10
algorithms = ['QL'] * n_episode

agents_list = Multiple_FMDP_Reco(n_episode, n_step, algorithms, 5, p, r)
```

```
agents_list.mean_epsilons
```

```
current_states = np.random.choice(range(3), 50)
current_states
```

```
agents_list.draw_action_matrix(current_states)
```
